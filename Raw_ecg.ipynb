```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("ecg_data.csv")

# Ensure the required columns exist
if "Time (ms)" not in df.columns or "Raw ECG" not in df.columns:
    raise ValueError("CSV file must contain 'Time (ms)' and 'Raw ECG values' columns.")

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(df["Time (ms)"], df["Raw ECG"], color="blue", label="ECG Signal")
plt.xlabel("Time (ms)")
plt.ylabel("Raw ECG")
plt.title("ECG Signal Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
